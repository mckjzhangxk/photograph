```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(-np.pi/2,np.pi/2,100)
y=np.sin(x)
plt.plot(x,y)
plt.title('y=sin(x)')
```

```python
x=np.linspace(-1,1,100)
y=np.arcsin(x)
plt.plot(x,y)
plt.title(r'$y=sin^{-1}(x),\frac{dy}{dx}=\frac{1}{\sqrt{1-x^2}}$')
```

```python
x=np.linspace(0,np.pi,100)
y=np.cos(x)
plt.plot(x,y)
plt.title('y=cos(x)')
```

```python
x=np.linspace(-1,1,100)
y=np.arccos(x)
plt.plot(x,y)
plt.title(r'$y=cos^{-1}(x),\frac{dy}{dx}=-\frac{1}{\sqrt{1-x^2}}$')
```

```python
eps=1e-6
x=np.linspace(-np.pi*0.4,np.pi*0.4)
y=np.tan(x)
plt.plot(x,y)
plt.title(r'$f(x)=tan(x),\frac{dy}{dx}=\frac{1}{(cosx)^2}$')
```

```python
x=np.linspace(-20,20,100)
y=np.arctan(x)
plt.plot(x,y)
plt.title(r'$f(x)=tan^{-1}(x),\frac{dy}{dx}=\frac{1}{1+x^2}$')
```

```python
secfunc=lambda x: 1/np.cos(x)
x=np.linspace(0,np.pi,100)
y=secfunc(x)
plt.plot(x,y)
plt.title(r'$f(x)=sec(x),\frac{dy}{dx}=\frac{sinx}{(cosx)^2},|sec(x)|>= 1$')
```

```python
xs=np.concatenate([y[50:],y[:50]])
ys=np.arccos(1/xs)
plt.plot(xs,ys)
plt.title(r'$f(x)=sec^{-1}(x),|x|>=1,\frac{dy}{dx}=\frac{1}{|x|\sqrt{x^2-1}}$')
```
