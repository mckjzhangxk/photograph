## A. Homogenous coordinates

In this task you need to find the intersection of two lines *$l_1$* and $l_2$. The line $l_1$ is defined by two points $x_1 = [98, 352]^T$ and $x_2 = [1548, 543]^T$. You can find the parameters of the line $l_2$ by computing the following $x_3$ and $x_4$ points.

The point $x_3$ is the intersection of the lines $m_1$ and $m_2$. The lines $m_1$ and $m_2$ are determined through the *Hesse normal form* with $\theta_1=35^{\circ}$, $d_1=1210$ and $\theta_2=138^{\circ}$, $d_2=978$. The point $x4$ can be found as an intersection point of lines $m_3$ and $m_4$.

The lines $m_3$and $m_4$ are given in the *intercept form*, where $x_0^3 = 1945$, $y_0^3 = 2761$ and $x_0^4 = -458$, $y_0^3 = 503$.

**Hints**:
- Hesse normal form: $cos(\theta)\cdot x + sin(\theta)\cdot y = d$ 
- Intercept form: $\frac{x}{x_0} + \frac{y}{y_0} = 1$

**Task**:
1. Find the intersection point $s$ of line $l_1$ and $l_2$ using **projective geometry** (homogenous coordinates). Print out the **euclidiean coordinates** of points $s$, $x_3$ and $x_4$. Finally, plot both lines l1 and l2 and verify your results.
<br> *(4 points)*

```python
# import all required modules
import numpy as np
import matplotlib as mlt
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# compute the coordinates of x_3 by using homogenous coordinates
theta1,d1=35*np.pi/180,1210,
theta2,d2=138*np.pi/180,978
M1=np.array([[np.cos(theta1),np.sin(theta1)],
             [np.cos(theta2),np.sin(theta2)],
            ])
x3=np.linalg.solve(M1,[d1,d2])
print(x3)
# 画图验证

fig,axes=plt.subplots(1)
t=np.linspace(200,300,100)
#m1 
Lm1,=axes.plot(t,(d1-t*M1[0][0])/M1[0][1],label='m1')
Lm1.set_color("#ff00ff")
#L2
Lm2,=axes.plot(t,(d2-t*M1[1][0])/M1[1][1],label='m2')
Lm2.set_color("#00ffff")
# #s
Sx3,=axes.plot(x3[0],x3[1],'o',label='x3')
Sx3.set_color("red")
fig.legend()
```

```python
# compute the coordinates of x_4 by using homogenous coordinates
M=np.array([
    [1/1945,1/2761],
    [-1/458,1/503]])
x4=np.linalg.solve(M1,[1,1])
```

```python
# Given x_1 and x_2 compute the line l_1
x1=[98,352]
x2=[1548,543]
ML1=np.array([x1,x2])
l1_coff=np.linalg.solve(ML1,[1,1])
print(l1_coff)
```

```python
# Given x_3 and x_4 compute the line l_2
ML2=np.array([x3,x4])
l2_coff=np.linalg.solve(ML2,[1,1])
print(l2_coff)
```

```python
# Given l_1 and l_2 compute the intersection points s
Ms=np.array([l1_coff,l2_coff])
s=np.linalg.solve(Ms,[1,1])
print(s)
```

```python
# print out the euclidean coordinates of s, x_3 and x_4
X3=np.concatenate([x3,[1]])
X4=np.concatenate([x4,[1]])
print(X3)
print(X4)
```

```python
# finally, plot both lines l1 and l2 and verify your results
fig,axes=plt.subplots(1)

t=np.linspace(0,80,100)
#L1 
L1,=axes.plot(t,(1-t*l1_coff[0])/l1_coff[1],label='l1')
L1.set_color("#ff00ff")
#L2
L2,=axes.plot(t,(1-t*l2_coff[0])/l2_coff[1],label='l2')
L2.set_color("#00ffff")
#s
S,=axes.plot(s[0],s[1],'o',label='s')
S.set_color("red")
fig.legend()
```

## C. DLT

```python
# X:(3,N), (x,y,1) in checkboard
# x:(3,N), (x,y,1) in sensor

#计算P 

def AX_Equal_Zeros(G):
    
    U,S,VT=np.linalg.svd(G)
    V=VT.T
    P_estimation=V[:,-1]
    return P_estimation

def DLT(X,x):
    G=[]
    n=X.shape[1]
    
    for i in range(n):
        Xi=X[:,i]
        xi,yi,_=x[:,i]
        ai=np.concatenate([Xi,np.zeros(len(Xi)),-xi*Xi])
        bi=np.concatenate([np.zeros(len(Xi)),Xi,-yi*Xi])
        G.append(ai)
        G.append(bi)
    G=np.array(G) 
    
    #G.shape=(2N,9)
    U,S,VT=np.linalg.svd(G)
    V=VT.T
    H=V[:,-1]
    H.shape=(3,len(Xi))
    return H

def Rx(d):
    c,s=np.cos(d/180*np.pi),np.sin(d/180*np.pi)
    M=[
        [1,0, 0],
        [0,c,-s],
        [0,s, c],
    ]
    return np.array(M)
def Ry(d):
    c,s=np.cos(d/180*np.pi),np.sin(d/180*np.pi)
    M=[
       [ c,0,s],
       [ 0,1,0],
       [-s,0,c],
    ]
    return np.array(M)
def Rz(d):
    c,s=np.cos(d/180*np.pi),np.sin(d/180*np.pi)
    M=[
        [c,-s,0],
        [s, c,0],
        [0, 0,1],
    ]
    return np.array(M)
# 返回3x4的矩阵 [R,-RX0]
def CalibrationM(a,b,c,X0):
    R=Rx(a)@Ry(b)@Rz(c)
    h=-R@X0
    M=np.zeros((3,4))
    M[:3,:3]=R
    M[:3,3]=-R@X0
    return M
```

```python
def UnCalibrationM(c,m,xh,yh,s):
    M=[
        [c,c*s,xh],
        [0,c*(1+m),yh],
        [0,0,1]
    ]
    return np.array(M)
```

```python
# 生成测试数据，R,X0,K
X0=np.array([1,2,3])
ax,ay,az=30,20,-241
R0=Rx(ax)@Ry(ay)@Rz(az)
R=CalibrationM(ax,ay,az,X0)

K=UnCalibrationM(-3,0.1,10,5,1)

print(R.shape)
print(K.shape)
P=K@R
```

```python
##生成 3D坐标X 与2D 坐标x 的对应 关系
# X：（4,n）,齐次坐标系
# x:  (3,n),齐次坐标系

n=6
X=np.random.rand(4,n)
X[3:]=1  #正则

px=P@X # 投影
x=px/px[2,:] #正则
```

```python

P_estimation=DLT(X,x)
```

```python
#验证正确性
np.testing.assert_almost_equal(P_estimation/P_estimation[-1][-1]-P/P[-1][-1],0)
```

```python
#还原X0
# P=[R -RX0]=[H h]
H=P_estimation[:,:3]
h=P_estimation[:,3]
X0_estimation=np.linalg.solve(H,-h)
np.testing.assert_almost_equal(X0_estimation-X0,0)
```

```python
#还原K,R
# print(Rz(180))
permute=np.array([[-1,0,0],[0,-1,0],[0,0,1]])

Q,OrthR=np.linalg.qr(np.linalg.inv(H))
R0_estimation=Q.T
# R0_estimation=permute.T@R0_estimation
# R0_estimation=-R0_estimation
np.testing.assert_almost_equal(R0,R0_estimation)

K_estimation=np.linalg.inv(OrthR)
K_estimation=K_estimation/K_estimation[-1][-1]
# K_estimation=K_estimation@permute
np.testing.assert_almost_equal(K,K_estimation)
```

###  len disort

```python
def drawLine(ax,data):
    for (xx,yy) in data:
        ax.plot(xx,yy,c='b')
def disort(data,q1,q2):
    result=[]
    for (xx,yy) in data:
        r2=xx**2+yy**2
        r4=r2*r2
        multi=1+q1*r2+q2*r4
        result.append((xx*multi,yy*multi))
    return result
```

```python
N=10
r=3
xx,yy=np.meshgrid(np.linspace(-3,3,N),np.linspace(-3,3,N))
lines=[]
for i in range(N):
    lines.append((xx[i],yy[i]))
    lines.append((xx[:,i],yy[:,i]))
    
fig,(ax1,ax2)=plt.subplots(1,2)
drawLine(ax1,lines)
ax1.set_aspect('equal', 'box')


# lines_disort=disort(lines,-0e-2,-33e-4)
q1,q2=-6e-3,-4e-4
lines_disort=disort(lines,q1,q2)
drawLine(ax2,lines_disort)
ax2.set_aspect('equal', 'box')

```

```python
# line to point
Xa=[]
for (xx,yy) in lines_disort:
    hs=np.vstack((xx,yy)).T
    Xa.append(hs)
Xa=np.array(Xa)
```

```python
#iterative invert solve
Xk=Xa
Xp=Xk

while True:   
    rs2=(Xk**2).sum(axis=2)
    rs4=rs2*rs2
    inv=1/(1+q1*rs2+q2*rs4)
    Xk=np.expand_dims(inv,2)*Xa
    if np.all(np.abs(Xp-Xk)<1e-5):
        break
    Xp=Xk
# print(Xa)
```

```python
fig,(ax1,ax2)=plt.subplots(1,2)
drawLine(ax1,lines)
ax1.set_aspect('equal', 'box')

line_estimate=[]
for d in Xk:
    line_estimate.append((d[:,0],d[:,1]))

drawLine(ax2,line_estimate)
ax2.set_aspect('equal', 'box')
```

### zhang's method

```python
def Vij(h,i,j):
    return np.array([
        h[0,i]*h[0,j],
        h[1,i]*h[0,j]+h[1,j]*h[0,i],
        h[2,i]*h[0,j]+h[2,j]*h[0,i],
        h[1,i]*h[1,j],
        h[1,i]*h[2,j]+h[1,j]*h[2,i],
        h[2,i]*h[2,j],
    ])
```

```python
#测试数据
K=UnCalibrationM(c=-2.4,m=0.5,xh=5,yh=15,s=1)
print(K)
```

```python
# 数学原理
R=CalibrationM(30,20,50,np.array([-3,-2,-1]))
# 取列的方法
RReduce=np.concatenate([R[:,0:2],R[:,3:]],axis=1)
P=K@RReduce
P=100*P

Pinv=np.linalg.inv(P)
h1=P[:,0]
h2=P[:,1]
B=np.linalg.inv(P@P.T)

print(h1.T@B@h1-h2.T@B@h2)
print(h1.T@B@h2)
```

```python
#测试数据
images=3

Xs=[]
xs=[]
Ps=[]


for i in range(images):
    # X必须来自一个平面
    X=np.array([
        [0,0,0,1,1,1],
        [0,1,2,0,1,2],
        [1,1,1,1,1,1]
    ])
#     #相机相对于 checkboard的旋转
    R=Rx(0)@Ry(np.random.randint(-10,10))@Rz(np.random.randint(-10,10))
#     #相机原点 相对于 checkboard的位移
    X0=-np.random.rand(3,1)*4
    
#     #checkout在 相机的 坐标转换
    RT=R.T
    Ra=np.concatenate([RT[:,0:2],-RT@X0],axis=1)
    
    
    P=K@ Ra
    
    #####tesing 
    BB=np.linalg.inv(P@P.T)
    h1=P[:,0]
    h2=P[:,1]
    
    np.testing.assert_almost_equal(0,h1.T@BB@h2)
    np.testing.assert_almost_equal(h1.T@BB@h1,h2.T@BB@h2)
    ##################
    
    Ps.append(P)
    
    x=P@X
    x=x/x[2,:]
    
    Xs.append(X)
    xs.append(x)
```

```python
# using zhang's method
G=[]
Ps_es=[]
for i in range(images):
    H=DLT(Xs[i],xs[i])
    Ps_es.append(H)
    aa=Vij(H,0,1)
    bb=Vij(H,0,0)-Vij(H,1,1)
    G.append(aa)
    G.append(bb)
G=np.array(G)
Bvector=AX_Equal_Zeros(G)

B=np.array([
    [Bvector[0],Bvector[1],Bvector[2]],
    [Bvector[1],Bvector[3],Bvector[4]],
    [Bvector[2],Bvector[4],Bvector[5]]
])
```

```python
np.linalg.eigvals(-B)
```

```python
# 验证DLT的正确性
for i in range(images):
    np.testing.assert_almost_equal(Ps_es[i]/Ps_es[i][-1,-1],Ps[i]/Ps[i][-1,-1])
```

```python
L=np.linalg.cholesky(B)
k_es=np.linalg.inv(L.T)
k_es=k_es/k_es[-1,-1]

permute=np.array([[-1,0,0],[0,-1,0],[0,0,1]])
k_es=k_es@permute
```
