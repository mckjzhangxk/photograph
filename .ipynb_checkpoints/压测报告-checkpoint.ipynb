```python
from collections import defaultdict
result=defaultdict(object)
# {"method": "joinRoom", "userId": "47410b520", "success": 1, "ts": 35257}

joinSuccess,joinFail,joinTs=0,0,[]
subSuccess,subFail,streams=0,0,[]
pubSuccess,pubFail,pubstreams=0,0,[]

with open('joinroom.txt','r') as f:
    for line in f.readlines():
        ii=line.index('stats')
        obj=json.loads(line[ii+5:])
        
        userId,method,ts,success=obj["userId"],obj["method"],obj["ts"],obj["success"]
        
        if userId not in result:
            result[userId]=defaultdict(object)
            
            result[userId]["streamsSuccess"]=0
            result[userId]["streamsFail"]=0
            result[userId]["sub.ts"]=0
            result[userId]["joinSuccess"]=0
            result[userId]["joinFail"]=0
        p=result[userId]

        if method=="joinRoom":
            if success==1:
                joinSuccess+=1
                joinTs.append(ts)
            else: 
                joinFail+=1
        elif method=="subStream":
            if success==1:
#                 streamId=obj["streamId"]
                subSuccess+=1
                streams.append(ts)
            else: 
                subFail+=1
        elif method=="pubStream":
            if success==1:
                pubSuccess+=1
                pubstreams.append(ts)
            else:
                pubFail+=1
```

```python
#加入房间的统计
fig, ax = plt.subplots()
ax.hist(np.array(joinTs)/1e6, bins=20, linewidth=0.5, edgecolor="white")

ax.set(xlim=(0, 9), xticks=np.linspace(0, 9, 10),ylim=(0, 1200), yticks=np.linspace(0, 500, 12))
plt.xlabel('delay')
plt.ylabel('#users')

plt.show()
print("JoinRoom 最小值=%f，最大值=%f，均值=%f，成功/失败=%d/%d"
      %(np.min(joinTs)/1e6,np.max(joinTs)/1e6,np.average(joinTs)/1e6,joinSuccess,joinFail))
```

```python
import numpy as np
import json
import matplotlib.pyplot as plt
def second(m):
    return m/1e6
def microSecond(m):
    return m*1e6
```

```python
mpcNum=9
N=2000

interval_in_micro=microSecond(120/2000)  #400 qps
join_broad_interval_in_micro=2*1000*1000
join_broad_broad_in_micro=1*1000*1000

def randomChooseMe():
    return np.random.randint(mpcNum)==0

def initialJoinRoomAndBroadcast():
    joinRoom=[]
    
    
    joinRoom_ts=np.random.randn(N)*interval_in_micro/5+interval_in_micro
    joinRoom_ts=np.cumsum(joinRoom_ts)
    
    broadcast_1=np.random.randn(N)*join_broad_interval_in_micro/5+join_broad_interval_in_micro
    broadcast_1=broadcast_1+joinRoom_ts
    
    broadcast_2=np.array([join_broad_broad_in_micro]*N) #等距间隔
    broadcast_2=broadcast_1+broadcast_2
    
    np.testing.assert_array_less(joinRoom_ts,broadcast_1)
    np.testing.assert_array_less(broadcast_1,broadcast_2)
    for i in range(N):
        if randomChooseMe():
            joinRoom.append(joinRoom_ts[i])
    
    joinRoom_ts= np.array(joinRoom)
    def Nf(t):
        r=np.where(joinRoom_ts<=t)[0]
        return len(r)
    
    return joinRoom_ts,np.concatenate((broadcast_1,broadcast_2)),Nf
```

```python
joinRoom_T,broadcast_T,nf=initialJoinRoomAndBroadcast()
personInMpc=len(joinRoom_T)
print("user/Mpc=%d,总人数=%d,#mpc=%d"%(personInMpc,N,mpcNum))
```

```python
#验证人数的正确性
xx=np.linspace(0,microSecond(60*2.1),10000) #30分钟
yy=[]
for x in xx:
    yy.append(nf(x))
plt.plot(second(xx),yy)
```

```python
C=24000/microSecond(1)
A=[0]
ts=np.concatenate(([0],joinRoom_T,broadcast_T))
ts=np.sort(ts)


broadcast_cnt=0
for i,t in enumerate(ts[1:]):    
    delta_t=t-ts[i]
    newbroadcast=not np.any(joinRoom_T==t)
    Aplus=max(A[i]-C*delta_t,0)+nf(t)*newbroadcast
#     Aplus=A[i]+nf(t)*newbroadcast
    A.append(Aplus)
    
    if newbroadcast:
        broadcast_cnt+=1
assert(broadcast_cnt==2*N)
```

```python
print('最大广播量=%d,A[-1]=%d,delta=%d'%((2*N*personInMpc,A[-1],2*N*personInMpc-A[-1])))
```

```python
def evalPayload(t):
    lowIndex,upperIndex=-1,-1
    vs=np.where(t<=ts)[0]
    if len(vs)==0:
        pass
    else:
        upperIndex=vs[0]
        lowIndex=upperIndex-1

    if upperIndex>=0 and lowIndex>=0:
        r=A[lowIndex]+(A[upperIndex]-A[lowIndex])*(t-ts[lowIndex])/(ts[upperIndex]-ts[lowIndex])
        if r<0:
            print(ts[lowIndex],t,ts[upperIndex])
#         r=1
    elif upperIndex==-1:
        r=max(A[-1]-C*(t-ts[-1]),0)
    else :
        r=0
    return r
    
# print(ts[a],ts[b])
# print(A[a],A[b])
```
