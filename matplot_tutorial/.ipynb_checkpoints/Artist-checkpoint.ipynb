```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

### Artist类别
* 基本类型：Line2D,Text,AxesImage,Rect
* 复合类型:Figuure,Axes,Subplot，Axis
* * Axes有很多helper方法，用于创建基本类型Artist,并且添加到复合类型中,比如plot,text，imshow
* * **Axes默认有 XAxis,YAxis2个对象，Axis也是复合对象，可以进行设置 尺度的显示**
* * Subplot是特殊的Axes, 表示grid 布局的axes
* * fig.add_axes([left, bottom, width, height]),可以添加任意布局的 axes

```python
fig=plt.figure()
axes1=fig.add_axes([0.2,0,0.4,0.4])
axes2=fig.add_axes([0.65,0,0.2,0.2])
axes3=fig.add_axes([0.90,0,0.1,0.1])

t = np.arange(0.0, 1.0, 0.01)
s = np.sin(2*np.pi*t)
line, = axes1.plot(t, s, color='blue', lw=2)
line, = axes2.plot(t, s, color='blue', lw=2)
line, = axes3.plot(t, s, color='blue', lw=2)



# axes1会把它包含的artist进行分类
axes1.lines[0]
# 你也可以随时移除某个 artist
axes1.lines[0].remove()
```

### example1
* subpolt与 一般的axes混合使用
* axes 中的 axis的操作
* label,title均为 Text

```python
fig = plt.figure()
# 这里作为一个参照
ax2 = fig.add_axes([0,0,1,1])

# 这里标识，把subplot的top 设置为0.8
fig.subplots_adjust(top=0.8)
ax1 = fig.add_subplot(211)
ylabel_text=ax1.set_ylabel('Voltage [V]')
title_text=ax1.set_title('A sine wave')
# 可以看到，label,title都是Text
# print(type(title_text))
# print(type(ylabel_text))
t = np.arange(0.0, 1.0, 0.01)
s = np.sin(2*np.pi*t)
line, = ax1.plot(t, s, color='blue', lw=2)

#axes 默认有XAxis,YAxis2个对象，这两个对象也是复合对象
ax1.xaxis.set_ticks([0,1])
ax1.xaxis.set_ticklabels(['0','T'])
ax1.yaxis.set_ticks([-1,0,1])
```

### Container
* figure: matplotlib.figure.Figure,是顶层容器
    * * add_subplot，.add_axes用于添加axes
    * * axes属性用于记录全部axes
    * * patch属性表示背景，是一个rect对象

```python
fig=plt.figure(figsize=(4,4))
fig.patch.set(color='yellow',alpha=0.7)

ax1=fig.add_subplot(211)
ax1.patch.set_color('red')
ax2=fig.add_axes([0,0,0.3,0.3])
ax2.patch.set_color('blue')
ax3=fig.add_axes([0.7,0,0.3,0.3])
ax3.patch.set_color('green')

for aa in fig.axes:
    aa.grid(True)
# ax3.remove()  不要直接增删fig.axes对象
```

* Axes container
* * Axes使用helper方法，创建 artist，并且把加入到axes中,并且返回给app，eg: hist,lines
* * lines,patches是2个重要的属性集合，分别对应add_line,add_patch方法
* * 包含2个重要的artist:xaxis and yaxis

```python
fig,(ax1,ax2)=plt.subplots(2,1)

line1,line2=ax1.plot([0,1,2],[0,1,2],'#ff0000',[0,1,2],[1,2,3],'#00ff00')
n,bins,bars=ax2.hist(np.random.randn(1000),bins=30,color='red')
# n: 每个bin对应的数量
# bins: 可以理解为bin的label
# bars:artist
```

### axes的 annotate方法

```python
fig, ax = plt.subplots()
arrow={"arrowstyle":"<-","color":'red'}
anotation=ax.annotate("",xy=(0,0),xytext=(0.3,0.4),arrowprops=arrow)

anotation.xy=(0.2,0.2)
anotation.set_position((0.6,0.6))
```
