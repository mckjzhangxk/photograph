## image in matplotlib

```python
#A. %matplotlib 是ipython的命令，是为了连上ipython GUI event loop，用于显示图片

#B. inline 是在notebook中画图，而且可以保证 plot之后的，命令不会影响到 之前的plot
#                  也就是说，plot是以command cell为单位的，不同的cell互不影响。

%matplotlib inline
```

```python
img=mpimg.imread('stinkbug.png') #这里是使用PILLOW的 imrad,RGB格式
```

```python
fig,ax=plt.subplots(1)
ax.imshow(img) #生成的是AxesImage对象
```

```python
# 单通道的显示，你可以指定不同的color map
# cmap:viridis,hot,nipy_spectral,mpl.colormaps()
lum_img = img[:, :, 0]
plotimg=plt.imshow(lum_img, cmap="viridis")
plotimg.set_cmap('nipy_spectral')
plt.colorbar()
```

```python
# 查看以下颜色分布的范围
_=plt.hist(img.ravel(),bins=256,range=[0,1],facecolor='k')
```

```python
# 增强对比度
# clim方法的2个参数 (l,u)
# eg: 图片中u,或者>u的pixvalue，都变成 源图片中的 最大pixel value对应的 亮度
# 图片中<=l的 pixvalue，都变成 源图片中的 最小 pixel value 对应的 亮度
fig,(ax1,ax2)=plt.subplots(1,2,layout='constrained')

pplot1=ax1.imshow(img[:,:,0],)  #原图最亮的 是黄色
fig.colorbar(pplot1,ticks=[0.1, 0.3, 0.5, 0.7], orientation='horizontal')

# >0.6的部分都显示黄色
pplot2=ax2.imshow(img[:,:,0],clim=(0,0.6))
# pplot2.set_clim(0,0.5)
fig.colorbar(pplot2, orientation='horizontal')
```
